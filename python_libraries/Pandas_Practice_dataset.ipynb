Adding / Removing columns using lambda

```python
Toyota = pd.read_csv("Toyota - Toyota.csv")
Toyota["_HP_"] = pd.to_numeric(Toyota["HP"] , errors="coerce")
Toyota["_HP_"]
```

```python
Toyota["HP"] = Toyota["_HP_"].apply(lambda x : "low" if x < 100 else ("high" if x > 300 else "medium"))
```

```python
Toyota.sample(5)
```

```python
def categories_cars(row):
    if row["_HP_"] < 100:
        return "Low"
    elif row["_HP_"] > 300:
        return "High"
    else:
        return "Medium"

Toyota["HP Categories"] = Toyota.apply(categories_cars , axis =1)
Toyota.head()
```

Pivot Table

```python
Toyota = Toyota.dropna()
```

```python
pivot = Toyota.pivot_table(
    columns="FuelType", 
    index="HP Categories", 
    values="Price", 
    aggfunc="median")
pivot
```
